```python
# create a custom exception class to handle empty queue condition during the operation of dequeue and display queue
class EmptyQueue(Exception):
    
    # constructor for custom class EmptyQueue
    def __init__(self, message):
        self.message = message
        
    def __str__(self):
        return self.message
```

```python
# create the class for Queue with associated operations called Enqueue and Dequeue
class Queue:
    
    # create a non public class called Node to represent or refer or template the atomic component of queue with data and a 
    # pointer pointing by default to None
    class _Node:
        
        def __init__(self, data):
            self._data = data
            self._pointer = None
    
    # constructor for queue to initialize the head to None
    def __init__(self):
        self._head = None
        
    # this method adds the node with the data that is passed as an argument to the end of the queue
    def _enqueue(self, data):

        # create a node with the passed data
        self._node = self._Node(data)
        
        # if the head of the queue is None then assign the new node as the head of the queue
        # otherwise, add the new node to the end of the queue
        if self._head is None:
            self._head = self._node
        else:
            temp = self._head
            while(temp._pointer is not None):
                temp = temp._pointer
            temp._pointer = self._node
        return 

    # this method removes the head node and assigns the node pointed by the head node under deletion as the new head node
    # this is because queue ADT (Abstract data structure) follows FIFO (First In First Out) behaviour
    def _dequeue(self):
        
        try:
            # if the head is None then raise the EmptyQueue exception
            # otherwise, print the head node and reassign the head node to its next node
            if self._head is None:
                raise EmptyQueue("Queue is empty")
            else:
                print("\nDequeuing", self._head._data, end="\n")
                self._head = self._head._pointer
        except EmptyQueue as e:
            return e
        except Exception as e:
            return e
    

    # this method prints the nodes in the queue in FIFO format
    def _show_queue(self):
        try:
            # if the head of the queue is empty then raise the EmptyQueue exception
            # otherwise, print it from the head to the last node of the queue
            if self._head is None:
                raise EmptyQueue("Queue is empty")
            else:
                temp = self._head
                while(temp is not None):
                    print(temp._data, end=" ")
                    temp = temp._pointer
        except EmptyQueue as e:
            return e
        except Exception as e:
            return e
```

```python
if __name__ == '__main__':
    # create a new Queue object
    queue = Queue()

    # add nodes with data to the queue
    queue._enqueue(10)
    queue._enqueue(20)
    queue._enqueue(30)
    queue._enqueue(40)

    # print the data items of the queueu to display the above inserted elements
    queue._show_queue()

    # remove the data node in the queue on the basis of FIFO
    queue._dequeue()

    # again, display the data items of the queue to see if the right item has been removed and the other nodes are intact
    queue._show_queue()
```
