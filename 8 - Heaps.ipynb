```python
# define a user defined exception class to handle empty heap
class EmptyHeap(Exception):
    
    # constructor for the EmptyHeap exception class
    def __init__(self, message):
        self.message = message
        
    # string dunder method for the class
    def __str__(self):
        return str(self.message)
```

```python
# define the class for heap
class Heap:
    
    # define the constructor for the heap with an empty list 'heap'
    def __init__(self):
        self._heap = [] 
        
    # method to return the parent index
    def _parent(self, index):
        return (index-1)//2
        
    # method to return the index of the left child
    def _left(self, index):
        return 2*index + 1
    
    # method to return the index of the right child
    def _right(self, index):
        return 2*index + 2
    
    # method to return the boolean value if the node has a left child or not
    def _has_left(self, index):
        return 2*index+1 < len(self._heap)
        
    # method to return the boolean value if the node has a right child or not
    def _has_right(self, index):
        return 2*index+2 < len(self._heap)
    
    # method to swap elements at given two indices    
    def _swap(self, parent_index, index):
        temp = self._heap[parent_index]
        self._heap[parent_index] = self._heap[index]
        self._heap[index] = temp
        
    # method to upheap the heap
    def _upheap(self, index):
        
        # identify the parent index of the node
        parent_index = (index-1)//2
        
        # while the node under focus is not root and the parent node is greater than the node 
        # focus swap the nodes and modify the indices
        while index > 0 and self._heap[parent_index] > self._heap[index]:
            self._swap(parent_index, index)
            index = parent_index
            parent_index = (index-1)//2
            
    # the above method to heapify the heap upon addition of a node can also be made recursive
     
    # this method if helpful at every remove min operation of heap
    def _downheap(self, index):
        
        # check if the node has a left child
        if self._has_left(index):
            left = self._left(index)
            child = left
            
            # check if the node also has a right child
            if self._has_right(index):
                right = self._right(index)
                
                # in case the node has both the children, identify the smallest child
                if self._heap[left] > self._heap[right]:
                    child = right
            
            # if the child is smaller than the parent then swap and call downheap again
            if self._heap[child] < self._heap[index]:
                self._swap(index, child)
                self._downheap(child)
    
    # this method adds a new node to the heap and also heapifies upon addition
    def _add_node(self, data):
        if len(self._heap) == 0:
            self._heap.append(data)
        else:
            self._heap.append(data)
            self._upheap(len(self._heap)-1)
            
    # this method removes the smallest node(the root node) and then heapifies with the 
    # rest of the nodes of the heap
    def _remove_min(self):
        try:
            if len(self._heap) == 0:
                raise EmptyHeap("Heap is Empty")
            else:
                self._heap[0] = self._heap[-1]
                self._heap.remove(self._heap[-1])
                self._downheap(0)
        except EmptyHeap as e:
            print(e)
    
    # this method prints all the nodes of the heap
    def _show_heap(self):
        try:
            if len(self._heap) == 0:
                raise EmptyHeap("The heap is empty")
            else:
                print(self._heap, end="\n")
        except EmptyHeap as e:
            print(e)       
```

```python
# adding the namespace for unit testing this module
if __name__ == '__main__':
    
    # create an object of the Heap class
    heap = Heap()
    
    # see if the EmptyHeap user defined exception works fine
    heap._show_heap()
    
    # add a few nodes to the heap
    heap._add_node(12)
    heap._add_node(17)
    heap._add_node(5)
    heap._add_node(18)
    heap._add_node(2)
    heap._add_node(1)    

    # print the nodes on the heap
    heap._show_heap()

    # remove the smallest node of the heap
    heap._remove_min()

    # print the nodes on the heap
    heap._show_heap()
```
