<h1>Single Linked List</h1>
<p>Let's create a class called Linked List with another sub class within it to template the Node having the attributes of data and pointer to the next node. Also, include various useful methods like adding node, removing a node, searching a node, printing all the nodes, finding the length of the linked list, etc.</p>

```python
class EmptyListException(Exception):
    
    # This is the constructor for the EmptyListException class
    def __init__(self, message):
        self.message = message
        
    def __str__(self):
        return self.message
```

```python
class LinkedList:
    
    # non-public class within class LinkedList to represent a node with data and pointer to the next node
    class _Node:
        
        # constructor for Node shall have non-public items called data and pointer where data is the element of the node and
        # pointer is the pointer/reference to the next node in the LinkedList which shall be None during instantiation
        def __init__(self, data):
            self._data = data
            self._pointer = None
    
    # constructor for LinkedList which sets the head node to None
    def __init__(self):
        self._head = None
        self._length = 0
        
    
    # add_last method that adds a node at the end of the Linked list
    def add_last(self, data):
        
        #create a node with the data passed as argument
        self._node = self._Node(data)
        self._length += 1
        
        # if the head node is empty then assign head pointer to pointer the new node that is created
        # otherwise, traverse through the Linkedlist till the last node and add the new node at the end
        if self._head is None:
            self._head = self._node
        else:
            temp = self._head
            while(temp._pointer!=None):
                temp = temp._pointer            
            temp._pointer = self._node
        return
    
    # add_first method that adds a node at the beginning of the Linked List
    def add_first(self, data):
        
        # create a node with the data passed as an argument
        self._node = self._Node(data)
        self._length += 1
        
        # if head of the linked list is empty then assign this new node as the head of the linked list 
        # else add the new node as the head of the linked list
        if self._head is None:
            self._head = self._node
        else:
            self._node._pointer = self._head
            self._head = self._node
        return   

    # this method return the presence of the data item in the linked list
    def search(self, data):
        try:
            if self._head is None:
                raise EmptyListException("Linked list is empty")
            else:
                temp = self._head
                while(temp is not None):
                    if temp._data == data:
                        return True
                    temp = temp._pointer
                
                return False
        except EmptyListException as e:
            return e
        except Exception as e:
            return e
        
    
    # this method adds a node after the node of the given data element
    def add_after(self, pos_data, data):
        try:
            if self._head is None:
                raise EmptyListException("Linked List is empty")
            else:
                self._node = self._Node(data)
                temp= self._head
                while(temp!=None):
                    if temp._data == pos_data:
                        self._node._pointer = temp._pointer
                        temp._pointer = self._node
                        self._length += 1
                        return 
                    temp = temp._pointer
            print("The positional data element is not found")        
            return 
        except EmptyListException as e:
            return e 
        except Exception as e:
            return e         
        
    # the below method removes the data node from the linked list
    def remove(self, data):
        try:
            if self._head is None:
                raise EmptyListException("Linked List is empty")
            else:
                if self._head._data == data:
                    if self._head._pointer is None:
                        self._head = None
                        return
                    else:
                        self._head = self._head._pointer
                        return
                else:
                    temp = self._head
                    while(temp._pointer is not None):
                        if temp._pointer._data == data:
                            temp._pointer = temp._pointer._pointer
                            return
                        temp = temp._pointer
                    return "The node to be removed is not found"
        except EmptyListException as e:
            return e
        except Exception as e:
            return e
    
    # returns the length of the linked list 
    def show_length(self):
        return self._length
    
    # print the data item of the nodes in the linked list
    def show_list(self):
        
        # if the head node is None then the list is empty.
        # Raise an exception or simply print it out for now
        if self._head is None:
            # print("Linked List is empty")
            raise EmptyListException("LinkedList is empty")
        else:
            temp = self._head
            while(temp != None):
                print(temp._data, end=" ")
                temp = temp._pointer
        return
```

```python
if __name__ == '__main__':

    # create an object of LinkedList
    linked_list = LinkedList()

    # printout the linkedlist to see an empty list output
    try:
        linked_list.show_list()
    except EmptyListException as e:
        print(e)
    except Exception as e:
        print(e)

    # add a few nodes to the linkedlist that you just created
    linked_list.add_first(10)
    linked_list.add_last(20)
    linked_list.add_after(20, 30)
    linked_list.add_last(40)

    # printout the linkedlist to see the nodes in the order they are inserted/added
    linked_list.show_list()

    # printout the length of the linked list
    print(linked_list.show_length())

    print(linked_list.search(20))
    print(linked_list.search(30))

    linked_list.remove(30)
    linked_list.show_list()
```
